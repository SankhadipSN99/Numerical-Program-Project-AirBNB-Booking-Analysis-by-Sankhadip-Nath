#### Chart - 5

Which are the top 10 visited places in each of the neighbouring groups?

````python
# Chart - 5 visualization code
#storing all the diffrent neighbourhood groups to diffrent data frames
brooklyn_group_df=df.loc[df['neighbourhood_group']== 'Brooklyn']
manhattan_group_df=df.loc[df['neighbourhood_group']== 'Manhattan']
Queens_group_df=df.loc[df['neighbourhood_group']== 'Queens']
Staten_Island_group_df=df.loc[df['neighbourhood_group']== 'Staten Island']
Bronx_group_df=df.loc[df['neighbourhood_group']== 'Bronx']
````

````python
# creating subplots to plot all the graphs together
fig, axes = plt.subplots(3,2, figsize=(42, 26))
plt.yticks(fontsize=30)
sns.countplot(ax=axes[0,0],y="neighbourhood", data=brooklyn_group_df, palette="icefire",
              order=brooklyn_group_df.neighbourhood.value_counts().iloc[:10].index).set_title('top 10 Brooklyn neighbourhood value count')

sns.countplot(ax=axes[0,1],y="neighbourhood", data=manhattan_group_df, palette="vlag",
              order=manhattan_group_df.neighbourhood.value_counts().iloc[:10].index).set_title('top 10 Manhattan neighbourhood value count')

sns.countplot(ax=axes[1,0],y="neighbourhood", data=Queens_group_df, palette="coolwarm",
              order=Queens_group_df.neighbourhood.value_counts().iloc[:10].index).set_title('top 10 Queens neighbourhood value count')

sns.countplot(ax=axes[1,1],y="neighbourhood", data=Staten_Island_group_df, palette="viridis",
              order=Staten_Island_group_df.neighbourhood.value_counts().iloc[:10].index).set_title('top 10 Staten_Island neighbourhood value count')

sns.countplot(ax=axes[2,0],y="neighbourhood", data=Bronx_group_df, palette="magma",
              order=Bronx_group_df.neighbourhood.value_counts().iloc[:10].index).set_title('top 10 Bronx neighbourhood value count')
````

##### 1. Why did you pick the specific chart?

The main motive to create this subplot is to get a clear visuals of the top 10 visited places specifically for all the neighbouring groups.

##### 2. What is/are the insight(s) found from the chart?

The top 3 neighbouring places in Brooklyn are Williamsburg, Stuyvesant and Bushwick respectively.


The top 3 neighbouring places in Manhattan are Harlem, Upper West Side and Hell's Kitchen respectively.

The top 3 neighbouring places in Queens are Astoria, Long Island City and Flushing respectively.

The top 3 neighbouring places in Staten_Island are St. George, Tompkinsville and Stapleton respectively.

The top 3 neighbouring places in Bronx are Kingsbridge, Fordham and Longwood respectively.

AirBNB should focus more on the top most polpular locations which are most preferred by the people in each neighbouring groups. In this way, they can attain a higher gain in their business.
